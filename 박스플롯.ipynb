```python
# 필요한 패키지 설치

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.rcParams['figure.dpi'] = 300
```

```python
# 데이터 불러오기
# https://www.kaggle.com/datasets/karthickveerakumar/startup-logistic-regression
df = pd.read_csv("/content/sample_data/50_Startups.csv")

# 데이터 샘플 확인
df.head()
```

```python
# Profit 변수로 기본 가로 세로 박스 플롯 시각화

# 세로 박스 플롯
plt.figure(figsize = (8, 6))
sns.boxplot(y = 'Profit', data = df)
plt.show()

# 가로 박스 플롯
plt.figure(figsize = (8, 2))
sns.boxplot(x = 'Profit', data = df)
plt.show()
```

```python
# State 구분에 따른 Profit 박스 플롯 시각화

plt.figure(figsize=(8,5))
sns.boxplot(x="State", y="Profit", data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# "State"별 "Profit"의 평균과 데이터 포인트를 포함한 박스 플롯을 그립니다.
sns.boxplot(x="State", y="Profit",
            showmeans=True,  # 박스 플롯 내에 평균값을 표시합니다.
            boxprops={'facecolor': 'None'},  # 박스의 색상을 투명으로 설정하여 안쪽을 비웁니다.
            data=df)

# "State"별 "Profit"의 데이터 포인트를 스트립 플롯으로 그립니다.
sns.stripplot(x='State', y='Profit',
              data=df,
              jitter=True,  # 데이터 포인트가 겹치지 않도록 흩어진 모습으로 표현합니다.
              marker='o',   # 데이터 포인트를 원형으로 표시합니다.
              alpha=0.5,    # 데이터 포인트의 투명도를 설정합니다.
              color='black' # 데이터 포인트의 색상을 검정색으로 지정합니다.
             )

plt.show()
```
